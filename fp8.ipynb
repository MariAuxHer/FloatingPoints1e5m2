```python
def compute_value(s, e, m):
    sign = (-1) ** s  # Calculate the sign (-1)^s

    exponent = 2 ** (e - 15)  # Calculate 2^(e-15)

    # Convert the two-bit mantissa to a fraction between 1 and 2
    if m == 0:
        mantissa = 0
    elif m == 1:
        
        mantissa = 0.25
    elif m == 2:
        mantissa = 0.5
    else:
        mantissa = 0.75

    value = sign * exponent * (1 + mantissa)  # Compute the final value

    return value
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the number of mantissa bits and exponent bits
mantissa_bits = 3
exponent_bits = 4

# Calculate the number of normal values
num_normals = 2 ** mantissa_bits * (2 ** exponent_bits - 1)

# Calculate the bias
bias = 2 ** (exponent_bits - 1) - 1

# Generate the normal values
normals = np.zeros(num_normals, dtype=float)
index = 0
for exponent in range(1, 2 ** exponent_bits):
    for mantissa in range(2 ** mantissa_bits):
        value = (1 + mantissa / (2 ** mantissa_bits)) * 2 ** (exponent - bias)
        normals[index] = value
        index += 1

# Create an array of x-axis values for plotting
x = np.arange(num_normals)

# Plot the normal values
plt.plot(x, normals, 'bo-')

# Set plot labels and title
plt.xlabel('Index')
plt.ylabel('Normal Values')
plt.title('FP8 Normal Values')

# Show the plot
plt.show()
```
